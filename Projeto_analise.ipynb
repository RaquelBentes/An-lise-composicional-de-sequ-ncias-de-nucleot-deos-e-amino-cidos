### 2. Reverso complementar

O reverso complementar é uma operação da biologia molecular e bioinformática em que é criada uma sequência que corresponde ao inverso da sequência original e as bases nitrogenadas são substituídas por suas complementares (adenina e timina, guanina e citosina)

Essa célula cria uma função chamada "reverso_complementar", nela, é criado o dicionário `complementar`, que apresenta essa correspondência, seguido da lista `seq_complementar` e da correção da sequência inserida pelo usuário, para possíveis erros de digitação. Após isso, é criada a condição para adicionar o complementar da base na nova lista, que é invertida e associada ao valor `reverso`. Por fim, todas as bases de `reverso` são unidas em uma string chamada `resultado`, que é chamada no return.

Tive dificuldade na parte de corrigir a sequência que o usuário iria apresentar, então senti a necessidade de utilizar a Inteligência Artificial (IA), unida aos materiais didáticos disponibilizados pelo professor Daniel Cassar para compreender como fazer isso. Os materiais estão disponíveis na referência, todavia perdi o link dessa interação com a IA.

```python
def reverso_complementar(sequencia):
    """Calcula o reverso complementar de uma sequência de DNA.

    A função recebe uma sequência de DNA (fita simples), remove espaços
    em branco, converte para letras maiúsculas e substitui cada base
    pela sua base complementar (A<->T, C<->G). Em seguida, inverte a
    ordem da sequência complementar resultante, simulando a fita
    antiparalela de DNA.

    Argumento:
        sequencia: Sequência de DNA contendo as bases A, T, C e G.
            Pode conter espaços, quebras de linha ou estar em letras
            minúsculas/maiúsculas, pois esses elementos são tratados
            internamente.

    Returns:
        str: O reverso complementar da sequência de entrada, caso todas
            as bases sejam válidas.
        str: Uma mensagem de erro no formato "Base inválida: X", caso
            seja encontrado algum caractere que não corresponda a uma
            base de DNA válida (A, T, C ou G).
    """
    complementar = {
        "A" : "T",
        "T" : "A",
        "G" : "C",
        "C" : "G"           
    }
    
    seq_complementar =[]
    seq_certa = "".join(sequencia.upper().split())
    for base in seq_certa:
        if base in complementar:
            seq_complementar.append(complementar[base])
        else:
            print(f'Base inválida identificada: {base}')
            return f'Base inválida: {base}' 
    
    reverso = seq_complementar [::-1]
    resultado = "".join(reverso)

    return resultado
```

### 3. Transcrição

É o processo no qual a célula copia a informação genética do DNA e a reescreve na forma de RNA. Para isso a função deve substituir o "T", que representa a base nitrogenada "timina" por "U", que representa a "uracila".

Essa célula apresenta uma função simples, chamada `sequencia_transcrita`. Nela, foi criado um dicionário que apresenta a relação das bases "T" e "U" e mantém as demais, uma string chamada `transcricao` e é padronizado a sequência inserida pelo usuário, removendo os espaços e colocando as letras em maiúsculo. Ao final ela percorre cada base da `seq_certa` e se estiver em `rna`, a sua base correspondente é adicionada à string `transcricao` e retorna a sequência transcrita.

```python
def sequencia_transcrita(sequencia):
    """Transcreve uma sequência de DNA em RNA mensageiro.

    A função recebe uma sequência de DNA, remove espaços em branco,
    converte para letras maiúsculas e substitui cada base pela base
    correspondente de RNA, trocando Timina (T) por Uracila (U) e
    mantendo Adenina (A), Guanina (G) e Citosina (C) inalteradas.

    Argumento:
        sequencia: Sequência de DNA contendo as bases A, T, C e G.
            Pode conter espaços, quebras de linha ou estar em letras
            minúsculas/maiúsculas, pois esses elementos são tratados
            internamente.

    Returns:
        str: A sequência de RNA transcrita a partir do DNA fornecido.
        
    Observações:
        Diferente da função `reverso_complementar`, esta função não
        verifica explicitamente bases inválidas: qualquer caractere
        que não esteja no dicionário `rna` (ou seja, diferente de
        A, T, G ou C) é simplesmente ignorado e não aparece no
        resultado final, sem gerar aviso ou erro.
    """
    rna = {
        "T": "U",
        "A": "A",
        "G": "G",
        "C": "C"
    }

    transcricao = ""
    seq_certa = "".join(sequencia.upper().split())

    for base in seq_certa:
        if base in rna:
            transcricao += rna[base] 

    return transcricao
```

### 4. Tradução

É o processo no qual a sequência do RNA mensageiro é decodificado em trios (trincas), que representam um aminoácidos que compõe o código da proteína.

Nessa célula é criada a função `traducao`, primeiro são criadas três estruturas: a string `proteina`, que vai armazenar os aminoácidos gerados, o dicionário `dic_codons` que vai guardar a relação entre códons e aminoácidos e a lista `codons`  que vai armazenar as trincas. Em seguida a sequência é dividida em códons com o `range` que faz a leitura de 3 em 3 posições, que são adicionados na lista. A segunda condição que aparece verifica o tamanho de cada códon na lista e se ele for menor que três o loop é interrompido. Após isso, há uma sequência de `if`s com a mesma estrutura que verificam quais códons são referentes a quais aminoácidos (para facilitar a escrita dessa relação, a IA foi utilizada, disponibilizando uma tabela, o link dessa conversa está disponível na sessão "Uso de Inteligência Artificial"). Ao encontrar o códon de parada (*) a função é interrompida e para unir todos os aminoácidos na string, `proteina +=dic_codons[codon]` é utilizado. O `else` é adicionado ao final caso haja possíveis erros ou um códon que não está presente. 

```python
def traducao(sequencia):
    """Traduz uma sequência de RNA mensageiro (mRNA) em uma sequência de aminoácidos.

    A função divide a sequência de entrada em códons (trincas de
    nucleotídeos) e converte cada códon no aminoácido correspondente,
    usando o código genético padrão (notação de uma letra). A tradução
    é interrompida automaticamente ao encontrar um códon de parada
    (UAA, UAG ou UGA), que é representado por "*" na proteína
    resultante.

    Argumento:
        sequencia (str): Sequência de RNA mensageiro, já em letras
            maiúsculas e contendo apenas as bases A, U, G e C (ou seja,
            o resultado de uma transcrição prévia, como a função
            `sequencia_transcrita`).
            
    Returns:
        str: A sequência de aminoácidos traduzida, em código de uma
            letra, incluindo um "*" ao final caso um códon de parada
            seja encontrado.
        str: Uma mensagem de erro no formato "Códon inválido:XXX",
            caso seja encontrado um códon de três bases que não
            corresponda a nenhum aminoácido ou códon de parada
            conhecido.

    Observações:
        - Códons incompletos no final da sequência (com menos de 3
          bases) são descartados silenciosamente, sem aviso.
        - A função espera uma sequência de RNA (com U), não de DNA.
          Sequências de DNA (com T) resultarão em "Códon inválido"
          para praticamente todos os códons, já que o dicionário
          interno só reconhece bases de RNA.
        - A função assume que a sequência já está em letras
          maiúsculas e sem espaços; ao contrário de
          `reverso_complementar` e `sequencia_transcrita`, ela não
          faz esse tratamento internamente.
    """
    proteina = ""
    dic_codons = {}
    codons = []
    
    for i in range(0, len(sequencia),3): 
        codons.append(sequencia[i:i+3]) 
        
    for codon in codons:
        if len(codon) < 3: 
            break
      
        if codon == "UUU" or codon == "UUC":
            dic_codons[codon] = "F"
        if codon == "UUA" or codon == "UUG" or codon == "CUU" or codon == "CUC" or codon == "CUA" or codon == "CUG":
            dic_codons[codon] = "L"
        if codon == "AUU" or codon == "AUC" or codon == "AUA":
            dic_codons[codon] = "I"
        if codon == "AUG":
            dic_codons[codon] = "M"
        if codon == "GUU" or codon == "GUC" or codon == "GUA" or codon == "GUG":
            dic_codons[codon] = "V"
        if codon == "UCU" or codon == "UCC" or codon == "UCA"or codon == "UCG"or codon == "AGU"or codon == "AGC":
            dic_codons[codon] = "S"
        if codon == "CCU" or codon == "CCC"or codon == "CCA"or codon == "CCG":
            dic_codons[codon] = "P"
        if codon == "ACU" or codon == "ACC"or codon == "ACA"or codon == "ACG":
            dic_codons[codon] = "T" 
        if codon == "GCU" or codon == "GCC"or codon == "GCA"or codon == "GCG":
            dic_codons[codon] = "A"
        if codon == "UAU" or codon == "UAC":
            dic_codons[codon] = "Y"
        if codon == "CAU" or codon == "CAC":
            dic_codons[codon] = "H"
        if codon == "CAA" or codon == "CAG":
            dic_codons[codon] = "Q"
        if codon == "AAU" or codon == "AAC":
            dic_codons[codon] = "N"
        if codon == "AAA" or codon == "AAG":
            dic_codons[codon] = "K"
        if codon == "GAU" or codon == "GAC":
            dic_codons[codon] = "D"
        if codon == "GAA" or codon == "GAG":
            dic_codons[codon] = "E"
        if codon == "UGU" or codon == "UGC":
            dic_codons[codon] = "C"
        if codon == "UGG":
            dic_codons[codon] = "W"
        if codon == "CGU" or codon == "CGC"or codon == "CGA"or codon == "CGG"or codon == "AGA"or codon == "AGG":
            dic_codons[codon] = "R"
        if codon == "GGU" or codon == "GGC"or codon == "GGA"or codon == "GGG":
            dic_codons[codon] = "G"
        if codon == "UAA" or codon == "UAG"or codon == "UGA":
            dic_codons[codon] = "*"

        if codon in dic_codons:
            if dic_codons[codon] == "*":
                proteina +=dic_codons[codon]
                break
            proteina +=dic_codons[codon]

        else:
            print(f'Códon inválido:{codon}')
            return f'Códon inválido:{codon}'
        
    return proteina
```

A função `expressao_genica` une as duas anteriores, aplicando na função `transcricao` o argumento recebido e esse resultado é aplicado na função `traducao`, retornando a sequência de aminoácidos que compõe a proteína da sequência de DNA ou RNA que o usuário colocar.

```python
def expressao_genica(dna):
    """Simula o processo de expressão gênica, transcrevendo e traduzindo uma sequência de DNA.

    Esta função une as etapas de transcrição e tradução em um único
    fluxo: primeiro converte a sequência de DNA em RNA mensageiro
    (chamando `sequencia_transcrita`) e, em seguida, traduz esse RNA
    em uma sequência de aminoácidos (chamando `traducao`).

    Argumento:
        dna (str): Sequência de DNA contendo as bases
            A, T, C e G. Pode conter espaços, quebras de linha ou
            estar em letras minúsculas/maiúsculas, pois isso é tratado
            internamente pela função `sequencia_transcrita`.

    Returns:
        str: A sequência de aminoácidos resultante da tradução do RNA,
            incluindo um "*" ao final caso um códon de parada seja
            encontrado.
        str: Uma mensagem de erro no formato "Códon inválido:XXX",
            caso a função `traducao` encontre um códon inválido durante
            o processo.
    """
    rna = sequencia_transcrita(dna)
    proteina = traducao(rna)
    
    return proteina    
```

### 5. Porcentagem do Conteúdo GC

É a proporção em porcentagem de bases nitrogenadas em uma molécula de DNA ou RNA formadas por guanina (G) ou citosina (C).

Na função `calcular_gc` é criada a lista vazia `bases_validas`, que posteriormente vai armazenar apenas as bases válidas; novamente `seq_certa` corrige a sequência do usuário. Em seguida, a sequência corrigida é percorrida, contanto que o número de aminoácidos seja diferente de zero, e se a base for válida, ela é adicionada na lista. A quantidade de bases "G" e "C" é contada e esse valor é dividido pela quantidade total, o que ao final é multiplicado por 100, para que o valor esteja em porcentagem.

```python
def calcular_gc(sequencia):
    """Calcula o conteúdo GC (porcentagem de Guanina e Citosina) de uma sequência de DNA.

    A função remove espaços em branco, converte a sequência para letras
    maiúsculas e filtra apenas as bases válidas de DNA (A, T, C, G),
    ignorando silenciosamente qualquer outro caractere. O conteúdo GC é
    calculado como a proporção de bases G e C em relação ao total de
    bases válidas encontradas.

    Args:
        sequencia (str): Sequência de DNA a ser analisada. Pode conter
            espaços, quebras de linha, letras minúsculas/maiúsculas e
            até caracteres inválidos, pois apenas as bases A, T, C e G
            são consideradas no cálculo.

    Returns:
        float: A porcentagem de bases G e C na sequência (entre 0 e
            100), caso existam bases válidas.
        str: "Sequência vazia", caso a sequência de entrada não
            contenha nenhum caractere após a remoção de espaços.
        str: "Não foram identificadas bases válidas", caso a sequência
            não contenha nenhuma base válida de DNA (A, T, C ou G).

   Observações:
        Caracteres inválidos presentes no meio de uma sequência
        válida (ex: "ATXGC") são simplesmente descartados e não
        afetam o cálculo nem geram aviso — apenas as bases A, T, C e G
        entram no denominador da porcentagem.
    """
    bases_validas = []
    seq_certa = "".join(sequencia.upper().split())
    
    if len(seq_certa) == 0:
        return "Sequência vazia"
        
    for b in seq_certa:
        if b == "A" or b == "T" or b =="C" or b == "G":
            bases_validas.append(b)
    if len(bases_validas) == 0:
        return "Não foram identificadas bases válidas"
        
    gc = bases_validas.count("G")+ bases_validas.count("C")
    porcentagem = (gc/len(bases_validas))* 100
    
    return porcentagem
```

### 6. Gráfico do Conteúdo GC

Nessa função, a sequência é dividida em janelas de 10 aminoácidos para permitir uma análise mais detalhada de cada posição. O eixo das abscissas representa esses trechos enquanto o das oordenadas representa a porcentagem do conteúdo em cada.

Essa célula se inicia com a importação e a assimilação de um apelido às bibliotecas que serão utilizadas ao longo dela, sendo elas "Pandas", que permite organizar as informações em forma de tabela, e "Matplotlib" que cria e salva o gráfico. Na função `grafico_gc` inicialmente são criadas duas listas vazias, `posicoes` e `gc_porcent` que vão compor a tabela e seus componentes serão utilizados no gráfico. O `for` dessa função percorre a sequência de acordo com o delimitado por `quantidade`, com isso os trechos são selecionados com 10 componentes; a linha `gc_porcent.append(calcular_gc(trecho))` insere esse trecho na função `calcular_gc` e o resultado é adicionado na lista `gc_porcent` e a posição desse trecho é guardada em `posicoes`. A linha seguinte utiliza o módulo `pd.DataFrame()` para criar a tabela com os valores das listas e o gráfico de linha é criado com o matplotlib, rotulado e salvo como imagem.

```python
import pandas as pd  
import matplotlib.pyplot as plt

def grafico_gc(sequencia):
    """Gera e salva um gráfico de linha do conteúdo GC ao longo da sequência de DNA.

    A função percorre a sequência usando uma janela deslizante de 10
    bases, calculando a porcentagem de GC (via `calcular_gc`) para
    cada trecho de 10 bases consecutivas. Os resultados são organizados
    em um DataFrame (posição inicial da janela x %GC) e plotados como
    um gráfico de linha, que é salvo em disco como um arquivo PNG.

    Argumentos:
        sequencia (str): Sequência de DNA a ser analisada. Pode conter
            espaços, quebras de linha e letras minúsculas/maiúsculas,
            pois isso é tratado internamente antes da análise.

    Returns:
        None: A função não retorna nenhum valor. Como efeito colateral,
            salva o gráfico gerado no arquivo "conteudo_gc.png" (no
            diretório de trabalho atual) e fecha a figura do matplotlib.

    Observações:
        - O tamanho da janela é fixo em 10 bases (variável `quantidade`),
          definido internamente e não configurável via parâmetro.
        - Se a sequência tiver menos de 10 bases válidas, o intervalo
          do `range` fica vazio (ou negativo), resultando em um
          DataFrame vazio e, possivelmente, um erro ao tentar plotar
          o gráfico.
        - Como `calcular_gc` pode retornar uma string (ex: "Sequência
          vazia" ou "Não foram identificadas bases válidas") em vez de
          um número, trechos problemáticos dentro da sequência podem
          inserir valores não numéricos na lista `gc_porcent`,
          potencialmente causando erro ou comportamento inesperado
          no gráfico gerado pelo pandas/matplotlib.
        - O arquivo é sempre salvo com o mesmo nome
          ("conteudo_gc.png"), sobrescrevendo qualquer gráfico
          anterior gerado pela função.
    """
    seq_usada = "".join(sequencia.upper().split())
    posicoes = []
    gc_porcent = []
    quantidade = 10
    
    for i in range(len(seq_usada) - quantidade + 1):
        trecho = seq_usada[i:i+quantidade]
        gc = calcular_gc(trecho)
        if isinstance(gc, str):
            continue
        gc_porcent.append(gc)
        posicoes.append(i+1)
        
    df = pd.DataFrame({ 
        "Posição": posicoes,
        "%GC": gc_porcent
    })
    grafico = df.plot.line("Posição","%GC",legend=False) 
    
    plt.xlabel("Posição na sequência") 
    plt.ylabel("% GC") 
    plt.title("Conteúdo GC ao longo da sequência") 

    plt.savefig ("conteudo_gc.png", bbox_inches = "tight")
    plt.close()    
```

```python
import customtkinter as ctk
from PIL import Image 

def reverso_complementar_callback():
    """Callback do botão "Reverso Complementar".

    Lê a sequência digitada no widget `entrada`, calcula seu reverso
    complementar (via `reverso_complementar`) e exibe o resultado na
    caixa de texto `resultado_box`, substituindo qualquer conteúdo
    anterior. Também esconde a imagem do gráfico de GC (caso esteja
    visível), já que essa ação não gera gráfico.

    Adicionais:
        - Limpa e atualiza o conteúdo de `resultado_box`.
        - Esconde o widget `label_imagem` (`pack_forget`).
        - Força a atualização da janela (`janela.update_idletasks`)."""
    valor = entrada.get()
    resultado = reverso_complementar(valor)
    
    resultado_box.delete("1.0", "end")
    resultado_box.insert("1.0", str(resultado)) 

    label_imagem.pack_forget()
    janela.update_idletasks()
    
def sequencia_transcrita_callback():
    """Callback do botão "Transcrição".

    Lê a sequência digitada no widget `entrada`, transcreve o DNA em
    RNA (via `sequencia_transcrita`) e exibe o resultado na caixa de
    texto `resultado_box`. Esconde a imagem do gráfico de GC, caso
    esteja visível.

    Adicionais:
        - Limpa e atualiza o conteúdo de `resultado_box`.
        - Esconde o widget `label_imagem` (`pack_forget`).
        - Força a atualização da janela (`janela.update_idletasks`)."""

    valor = entrada.get()
    resultado = sequencia_transcrita(valor)

    resultado_box.delete("1.0","end")
    resultado_box.insert("1.0", str(resultado))

    label_imagem.pack_forget()
    janela.update_idletasks()

def expressao_genica_callback():
    """Callback do botão "Tradução".

    Lê a sequência digitada no widget `entrada`, executa o processo
    completo de expressão gênica (transcrição + tradução, via
    `expressao_genica`) e exibe a proteína resultante na caixa de
    texto `resultado_box`. Esconde a imagem do gráfico de GC, caso
    esteja visível.

    Adicionais:
        - Limpa e atualiza o conteúdo de `resultado_box`.
        - Esconde o widget `label_imagem` (`pack_forget`).
        - Força a atualização da janela (`janela.update_idletasks`)."""
    valor = entrada.get()
    resultado = expressao_genica(valor)

    resultado_box.delete("1.0", "end")
    resultado_box.insert("1.0", str(resultado))

    label_imagem.pack_forget()
    janela.update_idletasks()

def calcular_gc_callback():
    """Callback do botão "Porcentagem do conteúdo GC".

    Lê a sequência digitada no widget `entrada`, calcula a
    porcentagem de GC (via `calcular_gc`) e exibe o resultado na
    caixa de texto `resultado_box`. Esconde a imagem do gráfico de
    GC, caso esteja visível.

    Adicionais:
        - Limpa e atualiza o conteúdo de `resultado_box`.
        - Esconde o widget `label_imagem` (`pack_forget`).
        - Força a atualização da janela (`janela.update_idletasks`)."""
    valor = entrada.get()
    resultado = calcular_gc(valor)

    resultado_box.delete("1.0", "end")
    resultado_box.insert("1.0", str(resultado))

    label_imagem.pack_forget()
    janela.update_idletasks()

def grafico_gc_callback():
    """Callback do botão "Gráfico do conteúdo GC".

    Lê a sequência digitada no widget `entrada`, gera o gráfico de
    conteúdo GC ao longo da sequência (via `grafico_gc`, que salva o
    arquivo "conteudo_gc.png" em disco), carrega essa imagem e a
    exibe no widget `label_imagem`, posicionando-a antes da caixa de
    resultados.

    Adicionais:
        - Sobrescreve e recarrega o arquivo "conteudo_gc.png".
        - Atualiza a imagem exibida em `label_imagem` e a torna
          visível (`pack`).
"""
    valor = entrada.get()
    resultado = grafico_gc(valor)

    imagem = ctk.CTkImage(light_image = Image.open("conteudo_gc.png"), dark_image = Image.open("conteudo_gc.png"), size = (500,360))
    
    label_imagem.configure(image = imagem) 
    label_imagem.image = imagem 
    label_imagem.pack(before = resultado_box, pady = 10, padx = 10)

modo = "dark"
def cores():
    """Alterna o tema visual da interface entre os modos claro e escuro.

    Verifica o valor da variável global `modo` e inverte o tema da
    aplicação usando `ctk.set_appearance_mode`, atualizando também o
    valor de `modo` para refletir o estado atual.

    Adicionais:
        - Altera a aparência global da janela (`ctk.set_appearance_mode`).
        - Modifica a variável global `modo`.
    """
    global modo
    if modo == "dark":
        ctk.set_appearance_mode("light") 
        modo = "light"
    else:
        ctk.set_appearance_mode("dark")
        modo ="dark"
    
    
# Configurações da Interface

# Geral:
ctk.set_appearance_mode("dark")
ctk.set_default_color_theme("green") 

janela = ctk.CTk() 
janela.title("Análise composicional de sequências de nucleotídeos e aminoácidos")
janela.geometry("900x650")

# Elementos:
titulo = ctk.CTkLabel(janela, text = "Análise composicional") 
titulo.pack()

subtitulo = ctk.CTkLabel(janela, text = "Apresente a sequência, selecione o que deseja e pronto!")
subtitulo.pack()

# Entrada:
entrada = ctk.CTkEntry(janela, width = 300 , placeholder_text = "Insira a sequência:") 
entrada.pack()

# Botões:

# Botão de modo escuro/claro:
Bce = ctk.CTkButton(janela, text = "Modo: claro/escuro", command = cores)
Bce.pack(anchor = "ne",pady = 0, padx = 0)

# Botão de reverso complementar:
Brc = ctk.CTkButton(janela, text = "Reverso Complementar", command = reverso_complementar_callback)
Brc.pack(pady=10, padx = 10)

# Botão de transcrição:
Bta = ctk.CTkButton(janela, text = "Transcrição", command = sequencia_transcrita_callback) 
Bta.pack(pady = 10, padx = 10)
 
# Botão de tradução:
Btu = ctk.CTkButton(janela, text = "Tradução", command = expressao_genica_callback)
Btu.pack(pady = 10, padx = 10)

# Botões do Conteúdo GC:
    #Porcentagem:
Bpgc = ctk.CTkButton(janela, text = "Porcentagem do conteúdo GC", command = calcular_gc_callback)
Bpgc.pack(pady = 10, padx = 10)
    #Gráfico:
Bggc = ctk.CTkButton(janela, text = "Gráfico do conteúdo GC", command = grafico_gc_callback)
Bggc.pack(pady = 10, padx = 10)

#Label da Imagem
label_imagem = ctk.CTkLabel (janela, text = "")
label_imagem.pack(pady = 10, padx = 10)

# Caixa de resultados:
resultado_box = ctk.CTkTextbox(janela, height = 300, width = 600)
resultado_box.pack(pady = 10)

janela.mainloop() 
```
